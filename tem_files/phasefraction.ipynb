This code is for calculating the phase fractions of database and TEM images, which is the percent of the pixels that are black (after converting to binary).

```python
import cv2
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import shutil
```

```python
tem = "tem_images"
data = os.path.join("data","spinodal_images","crop_images")
```

```python
database = []
for path in os.listdir(data):
    sample = os.path.join(data,path)
    image = cv2.imread(sample,)
    image = np.array(tf.image.rgb_to_grayscale(image))
    thresh, image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
    image = image.flatten()
    total = image.shape[0]
    count = np.count_nonzero(image)
    frac = 1 - (count / total)
    database.append(frac)
```

```python
experimental = []
good_images = []
for path in os.listdir(tem):
    sample = os.path.join(tem,path)
    image = cv2.imread(sample)
    image = np.array(tf.image.rgb_to_grayscale(image))
    thresh, image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
    # image = np.array(tf.bitwise.invert(image), dtype = np.uint8)
    image = cv2.medianBlur(image, 3)
    imagef = image.flatten()
    total = imagef.shape[0]
    count = np.count_nonzero(imagef)
    frac = 1 - (count / total)
    experimental.append(frac)
    if not os.path.exists("good_tem_images"):
        os.mkdir("good_tem_images")
    if frac < 0.4:
        cv2.imwrite(os.path.join("good_tem_images", path), image)
```

```python
db = np.array(database)
exp = np.array(experimental)

print(db.mean(), db.std())

print(exp.mean(), exp.std())
```
